```python
import pandas as pd
import matplotlib.pyplot as plt

# Читаем CSV файл с результатами
df = pd.read_csv("timings.csv")

# Группируем данные по алгоритму
for algo, sub in df.groupby("algorithm"):
    sub = sub.sort_values("N")  # Сортируем по размеру N
    plt.plot(sub["N"], sub["time_us"], label=algo)

# Подписываем оси и добавляем легенду
plt.xlabel("Размер массива N")
plt.ylabel("Время (микросекунды)")
plt.title("Зависимость времени от размера массива")
plt.legend()
plt.grid(True)

# Сохраняем график в файл
plt.savefig("sorting_times.png", dpi=200)

# Показываем график
plt.show()
```
